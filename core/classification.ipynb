# Classification

Differentiating 3 flower types based on their features

```python
import tensorflow as tf
import pandas as pd
import numpy as np
```

```python
train_path = tf.keras.utils.get_file("iris_trainig.csv","https://storage.googleapis.com/download.tensorflow.org/data/iris_training.csv")
test_path = tf.keras.utils.get_file("iris_test.csv","https://storage.googleapis.com/download.tensorflow.org/data/iris_test.csv")
```

```python
CSV_COLUMN_NAMES = ['SepalLength', 'SepalWidth', 'PetalLength','PetalWidth','Species']
SPECIES = ['Setosa','Versicolor', 'Virginica']

train = pd.read_csv(train_path, names=CSV_COLUMN_NAMES, header=0)
test = pd.read_csv(test_path, names=CSV_COLUMN_NAMES, header=0)
train_y = train.pop('Species')
test_y = test.pop('Species')

train_y = pd.get_dummies(train_y, columns=['Species'], drop_first=False)
test_y = pd.get_dummies(test_y, columns=['Species'], drop_first=False)


train.head()
```

```python
train_y.head()
```

# Explore

```python
print('train:',train.shape,
      'test:', test.shape)
```

# training

```python
model = tf.keras.Sequential([
  tf.keras.layers.Dense(10, activation=tf.nn.relu, input_shape=(4,)),  # input shape required
  tf.keras.layers.Dense(10, activation=tf.nn.relu),
  tf.keras.layers.Dense(3, activation='softmax', name='output')
])

# Compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

model.fit(train, train_y, verbose=2, batch_size=5, epochs=200)
```

```python
results = model.evaluate(test, test_y)
```

```python

print('Final test set loss: {:4f}'.format(results[0]))
print('Final test set accuracy: {:4f}'.format(results[1]))
```

```python
predictions = pd.DataFrame(model.predict(test))
```
